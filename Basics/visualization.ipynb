```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting examples
x = [10,20,30]
y = [100,200,300]
plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
```

```python
# scatter plots using pyplot
plt.scatter(housing["rooms"],housing["price"])
```

```python
# xlim and ylim are used to set the lower and upper limits

plt.plot(housing["rooms"],housing["price"])
plt.xlim(0,3)
plt.ylim(0,350)
plt.title("Plotting Example")
plt.show()
```

```python
df = pd.read_csv("C:\\Users\\KARAN S07\\Desktop\\Final Year\\Deep Learning\\TF-Course-Notebooks\\TF_Course_Notebooks\\DATA\\heart.csv")
```

```python
df.head()
```

```python
# distplot describes the distribution of data in the form of a histogram
```

```python
sns.distplot(df["age"])
plt.show()

# the line signifies Kernel density estimation, a probablitic distribution.
```

```python
# to turn off the kde line
# each column acts as a bin, depicting the count falling under that range
sns.distplot(df["age"],kde=False)
plt.show()
```

```python
df.head()
```

```python
# to count the number of data points in a particular column, in this example sex
sns.countplot(x="sex",data=df)
plt.show()
```

```python
# to add a categorical feature use hue parameter
```

```python
sns.countplot(x="cp",data=df,hue="sex")
```

### Box plots 

Box plots show the distribution across various categories




```python
df.head()
```

```python
# for box plot x should be a categorical variable, and y is a continuous variable
sns.boxplot(x= "sex",y="age",data=df)
```

```python
sns.boxplot(x="target",y="thalach",data=df,hue="sex")
```

### Scatter plots

```python
# specifying size to points for instance age, bigger the point higher the age
sns.scatterplot(x="chol",y="trestbps",data=df,hue="sex",size="age")
```
